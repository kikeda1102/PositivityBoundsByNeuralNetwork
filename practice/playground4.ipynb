```python
import numpy as np
import math
import pandas as pd
from itertools import product
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
# from tensorflow.keras.optimizers import Adam
# M1/m2 Mac上での警告回避
from tensorflow.compat.v1.keras.optimizers import Adam as LegacyAdam

tf.config.run_functions_eagerly(True) # eager executionを有効に
tf.data.experimental.enable_debug_mode() # なんか警告でたのでデバッグモードを有効に
```

```python
# 2in 1outのNN toy model
```

```python
# サンプルデータ生成
np.random.seed(42)
X_train = np.random.rand(100, 2)  # 100サンプル, 2次元の特徴
y_train = X_train[:, 0] + 2 * X_train[:, 1]  # 簡単な線形関係
```

```python
X_train
```

```python
# モデルの構築
model = Sequential([
    Dense(4, activation='relu', input_dim=2),  # 入力層（2次元 -> 4次元）
    Dense(1, activation='linear')  # 出力層（1次元）
])
```

```python
# モデルのコンパイル
model.compile(optimizer='adam', loss='mean_squared_error')

# モデルの概要表示
model.summary()

# モデルの学習
model.fit(X_train, y_train, epochs=50, verbose=1)
```

```python
# サンプルデータ生成
X_test = np.random.rand(10, 2)  # 10サンプル, 2次元の特徴
y_test = X_test[:, 0] + 2 * X_test[:, 1]  # 簡単な線形関係
```

```python
model.predict(X_test)
```
